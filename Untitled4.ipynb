```python
import os
import docx
import nltk
import re
import string
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from textblob import TextBlob

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import LatentDirichletAllocation

from wordcloud import WordCloud
from transformers import pipeline

# 📦 NLTK Resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

from docx import Document

def create_sample_docx(file_name, content):
    doc = Document()
    doc.add_heading("Sample Document", level=1)
    doc.add_paragraph(content)
    doc.save(file_name)

create_sample_docx("doc1.docx", "Natural Language Processing is a fascinating field of AI.")
create_sample_docx("doc2.docx", "Deep Learning is used in NLP to understand human language.")

# 🧠 Load and Read .docx
def read_docx(file_path):
    with open(file_path, 'rb') as f:
        doc = docx.Document(f)
    return ' '.join([para.text for para in doc.paragraphs])

doc1_text = read_docx("doc1.docx")
doc2_text = read_docx("doc2.docx")

# 🧹 Preprocessing
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    text = re.sub(r"\d+", "", text)
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in stop_words]
    return " ".join(tokens)

doc1_clean = preprocess(doc1_text)
doc2_clean = preprocess(doc2_text)

# 🔹 Tokenization
tokens_doc1 = word_tokenize(doc1_clean)
tokens_doc2 = word_tokenize(doc2_clean)

# 🔹 POS Tagging
pos_doc1 = pos_tag(tokens_doc1)
pos_doc2 = pos_tag(tokens_doc2)

# ❤️ Sentiment Analysis
sentiment_doc1 = TextBlob(doc1_clean).sentiment
sentiment_doc2 = TextBlob(doc2_clean).sentiment

# 📊 TF-IDF + Cosine Similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([doc1_clean, doc2_clean])
similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

# ☁️ WordClouds
wordcloud1 = WordCloud(width=800, height=400, background_color='white').generate(doc1_clean)
wordcloud2 = WordCloud(width=800, height=400, background_color='white').generate(doc2_clean)

# 📈 TF-IDF Top Keywords
tfidf_df = np.mean(tfidf_matrix.toarray(), axis=0)
top_keywords = dict(sorted(zip(vectorizer.get_feature_names_out(), tfidf_df), key=lambda x: x[1], reverse=True)[:10])

plt.figure(figsize=(8,4))
sns.barplot(x=list(top_keywords.values()), y=list(top_keywords.keys()))
plt.title("Top TF-IDF Keywords")
plt.xlabel("Score")
plt.tight_layout()
plt.show()

# 🔍 LDA Topic Modeling
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(tfidf_matrix)

def display_topics(model, feature_names, n_words):
    for idx, topic in enumerate(model.components_):
        print(f"\nTopic {idx + 1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_words - 1:-1]]))

display_topics(lda, vectorizer.get_feature_names_out(), 5)

# 🤖 GPT-2 Text Generation
print("\n--- GPT-2 Text Generation ---")
generator = pipeline("text-generation", model="distilgpt2")
output = generator(doc1_clean, max_length=50, num_return_sequences=1)
print("Generated Text:", output[0]['generated_text'])

# 🔮 Zero-Shot Classification
print("\n--- Zero-Shot Classification ---")
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
candidate_labels = ["Education", "News", "Product Review", "Spam", "Story"]
result = classifier(doc2_clean, candidate_labels)
print("Predicted Label:", result['labels'][0], "→", f"{result['scores'][0]*100:.2f}% confidence")

# 📋 Output Summary
print("\n--- TOKENIZATION ---")
print("Doc1 Tokens:", tokens_doc1[:10])
print("Doc2 Tokens:", tokens_doc2[:10])

print("\n--- POS TAGGING ---")
print("Doc1 POS:", pos_doc1[:10])
print("Doc2 POS:", pos_doc2[:10])

print("\n--- SENTIMENT ---")
print(f"Doc1 → Polarity: {sentiment_doc1.polarity:.2f}, Subjectivity: {sentiment_doc1.subjectivity:.2f}")
print(f"Doc2 → Polarity: {sentiment_doc2.polarity:.2f}, Subjectivity: {sentiment_doc2.subjectivity:.2f}")

print("\n--- COSINE SIMILARITY ---")
print(f"Similarity between Doc1 & Doc2: {similarity:.2f}")

# 📊 WordClouds
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.imshow(wordcloud1, interpolation='bilinear')
plt.title("Document 1 Word Cloud")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.title("Document 2 Word Cloud")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Load dataset
df = pd.read_csv('Train_data.csv')

# 2. Select text columns
text_columns = ['protocol_type', 'service', 'flag']

# 3. Preprocessing
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from textblob import TextBlob
import re
import string

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in stop_words]
    return " ".join(tokens)

# Apply preprocessing and tokenization to each column
for col in text_columns:
    df[f'{col}_clean'] = df[col].apply(preprocess)
    df[f'{col}_tokens'] = df[f'{col}_clean'].apply(word_tokenize)
    df[f'{col}_pos'] = df[f'{col}_tokens'].apply(pos_tag)
    # Sentiment (not meaningful for categorical short text, but included)
    df[f'{col}_sentiment'] = df[f'{col}_clean'].apply(lambda x: TextBlob(x).sentiment)

# 4. Combine columns for document-level analysis
df['combined_text'] = df[text_columns].fillna('').agg(' '.join, axis=1)
df['combined_clean'] = df['combined_text'].apply(preprocess)

# 5. TF-IDF and Cosine Similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['combined_clean'])

# Cosine similarity between first two rows
sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
print("Cosine similarity between row 0 and 1:", sim)

# 6. Word Cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(df['combined_clean'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for protocol_type + service + flag")
plt.show()

# 7. LDA Topic Modeling
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(tfidf_matrix)

def display_topics(model, feature_names, n_words):
    for idx, topic in enumerate(model.components_):
        print(f"\nTopic {idx + 1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_words - 1:-1]]))

display_topics(lda, tfidf.get_feature_names_out(), 5)

# 8. Show tokenization, POS, sentiment for each column (first 3 rows)
for col in text_columns:
    print(f"\n--- {col.upper()} ---")
    for i in range(3):
        print(f"Row {i}:")
        print("  Clean:", df[f'{col}_clean'].iloc[i])
        print("  Tokens:", df[f'{col}_tokens'].iloc[i])
        print("  POS:", df[f'{col}_pos'].iloc[i])
        print("  Sentiment:", df[f'{col}_sentiment'].iloc[i])

# 9. (Optional) Zero-shot classification - not very meaningful for these columns, but you can do:
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
result = classifier(df['combined_clean'][0], ["network", "attack", "normal", "service"])
print("hi")
print(result)
```
